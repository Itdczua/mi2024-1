### **Теоретичний матеріал для тесту з машинного навчання**

---

#### **1. Розділення даних: тренувальний, валідаційний, тестовий набори**
- **Тренувальний набір:** Використовується для навчання моделі. Модель аналізує ці дані, щоб знайти залежності між вхідними та вихідними значеннями.
- **Валідаційний набір:** Допомагає налаштувати модель, зокрема її гіперпараметри, і уникнути перенавчання.
- **Тестовий набір:** Призначений для оцінки якості моделі на нових, невідомих даних. Це фінальна перевірка продуктивності моделі.

---

#### **2. Методи оцінки моделей**
##### **Класифікаційні моделі:**
- **Accuracy:** Відсоток правильно передбачених прикладів серед усіх.
- **Precision:** Визначає, яка частка позитивних передбачень правильна.
- **Recall:** Показує, яка частка реальних позитивних випадків правильно виявлена.
- **F1-Score:** Середнє значення Precision та Recall, використовується на незбалансованих даних.
- **ROC-AUC:** Відображає співвідношення істинно позитивних і хибно позитивних передбачень.

##### **Регресійні моделі:**
- **Mean Squared Error (MSE):** Середній квадрат помилок між передбаченнями і реальними значеннями.
- **Mean Absolute Error (MAE):** Середнє абсолютне значення помилок.
- **R²:** Показує, яка частка варіації у даних пояснюється моделлю.

---

#### **3. Гіперпараметри моделі**
- **Що це:** Це параметри, які задаються перед початком навчання моделі, наприклад, кількість дерев у Random Forest.
- **Як налаштувати:**
  - **Grid Search:** Перебір усіх можливих комбінацій гіперпараметрів.
  - **Random Search:** Випадковий перебір гіперпараметрів.
  - **Manual Search:** Ручний підбір параметрів на основі досвіду.
- **Приклад:** Налаштування глибини дерева рішень або кількості нейронів у шарі нейронної мережі.

---

#### **4. Попередня обробка даних**
- **Очищення даних:** Видалення або заповнення пропущених значень, видалення дублікатів, обробка аномалій.
- **Масштабування даних:**
  - **Standardization:** Приведення до стандартного нормального розподілу.
  - **Normalization:** Приведення значень до діапазону [0, 1].
- **Перетворення категоріальних змінних:**
  - **One-Hot Encoding:** Перетворення категоріальних змінних на числові через створення додаткових колонок.
  - **Label Encoding:** Привласнення категоріям числових значень.

---

#### **5. Крос-валідація**
- **Що це:** Метод оцінки моделі шляхом розбиття даних на кілька частин (folds). Модель навчається на \(k-1\) частинах, а перевіряється на залишеній.
- **Чому це важливо:** Забезпечує надійну оцінку моделі, знижуючи ризик залежності від одного поділу даних.

---

#### **6. Методи запобігання перенавчанню**
- **Регуляризація:** Додає штраф за складність моделі, щоб уникнути запам’ятовування даних.
- **Dropout:** У випадковому порядку виключає нейрони у нейронних мережах під час навчання.
- **Збільшення даних:** Додавання нових прикладів до тренувального набору.
- **Крос-валідація:** Перевірка моделі на різних підмножинах даних.

---

#### **7. Методи ансамблю**
- **Random Forest:** Комбінація багатьох дерев рішень для покращення точності.
- **Gradient Boosting:** Побудова моделей послідовно з акцентом на помилки попередніх.
- **Stacking:** Комбінація кількох моделей, щоб покращити результати.

---

#### **8. Зменшення розмірності**
- **PCA (Principal Component Analysis):** Метод, який зменшує кількість змінних, зберігаючи максимум інформації.
- **t-SNE:** Алгоритм для візуалізації даних у 2D або 3D просторах.
- Зменшення розмірності допомагає уникнути "прокляття розмірності" та знижує складність моделі.

---

#### **9. Метрики для незбалансованих даних**
- **Precision, Recall і F1-Score** часто є кращими за Accuracy.
- **ROC-AUC** дозволяє оцінити продуктивність моделі на різних порогах.

---

#### **10. Технічні інструменти**
- **Scikit-learn:** Для роботи з моделями машинного навчання.
- **Pandas:** Для роботи з даними у форматі таблиць.
- **NumPy:** Для обчислень із великими масивами даних.
- **Matplotlib і Seaborn:** Для візуалізації.
